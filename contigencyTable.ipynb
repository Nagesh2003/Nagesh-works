```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# =============================================
# Dataset 1: NON-SIGNIFICANT Relationship
# Gender vs. Beverage Preference
# =============================================
data1 = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Beverage': ['Coffee', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Coffee', 'Tea', 'Tea']
}
df1 = pd.DataFrame(data1)

cont_table1 = pd.crosstab(df1['Gender'], df1['Beverage'])
print("Dataset 1 - No Significant Relationship:")
print(cont_table1)
print("\n")

chi2_1, p1, dof1, expected1 = chi2_contingency(cont_table1)
print(f"Chi2 Statistic: {chi2_1:.4f}")
print(f"P-value: {p1:.4f}")
print("Expected Frequencies:")
print(expected1)
print("\n" + "="*50 + "\n")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(cont_table1, annot=True, cmap='Blues', fmt='d')
plt.title("No Significant Relationship\n(Gender vs. Beverage)")
plt.tight_layout()
plt.show()
# =============================================
# Dataset 2: SIGNIFICANT Relationship
# Age Group vs. Movie Preference
# =============================================
data2 = {
    'Age': ['Young', 'Young', 'Old', 'Old', 'Young', 'Young', 'Old', 'Old',
            'Young', 'Old', 'Old', 'Old', 'Young', 'Young'],
    'Movie': ['Action', 'Comedy', 'Comedy', 'Drama', 'Action', 'Action',
              'Drama', 'Drama', 'Action', 'Drama', 'Drama', 'Comedy', 'Action', 'Comedy']
}
df2 = pd.DataFrame(data2)

cont_table2 = pd.crosstab(df2['Age'], df2['Movie'])
print("Dataset 2 - Significant Relationship:")
print(cont_table2)
print("\n")

chi2_2, p2, dof2, expected2 = chi2_contingency(cont_table2)
print(f"Chi2 Statistic: {chi2_2:.4f}")
print(f"P-value: {p2:.4f}")
print("Expected Frequencies:")
print(expected2)
print("\n" + "="*50 + "\n")

# =============================================
# Visualization
# =============================================


plt.subplot(1, 2, 2)
sns.heatmap(cont_table2, annot=True, cmap='Reds', fmt='d')
plt.title("Significant Relationship\n(Age vs. Movie Preference)")

plt.tight_layout()
plt.show()

# =============================================
# Manual Chi-Square Calculation (Dataset 1)
# =============================================

# Contingency Table (Observed Frequencies):
#            Coffee   Tea   Total
#   Male        3      1      4
#   Female      3      1      4
#   Total       6      2      8

# Step 1: Expected Frequencies = (Row Total * Column Total) / Grand Total
# Expected for Male-Coffee = (4*6)/8 = 3.0
# Expected for Male-Tea    = (4*2)/8 = 1.0
# Expected for Female-Coffee = (4*6)/8 = 3.0
# Expected for Female-Tea = (4*2)/8 = 1.0

# Step 2: Chi-square Formula: Σ[(O - E)^2 / E]
# (3-3)^2 / 3 = 0
# (1-1)^2 / 1 = 0
# (3-3)^2 / 3 = 0
# (1-1)^2 / 1 = 0

# Total chi-square = 0
# Degrees of freedom = (2-1)*(2-1) = 1
# At α = 0.05, critical value from chi-square table = 3.841
# Since 0 < 3.841, result is NOT significant.

# Conclusion: ❌ No association between Gender and Beverage.

# =============================================
# Manual Chi-Square Calculation (Dataset 2)
# =============================================

# Contingency Table (Observed Frequencies):
#          Action  Comedy  Drama  Total
# Young      5       3       0      8
# Old        0       2       4      6
# Total      5       5       4     14

# Step 1: Expected Frequencies = (Row Total * Column Total) / Grand Total
# Young-Action = (8*5)/14 ≈ 2.857
# Young-Comedy = (8*5)/14 ≈ 2.857
# Young-Drama = (8*4)/14 ≈ 2.286
# Old-Action = (6*5)/14 ≈ 2.143
# Old-Comedy = (6*5)/14 ≈ 2.143
# Old-Drama = (6*4)/14 ≈ 1.714

# Step 2: Chi-square Formula: Σ[(O - E)^2 / E]
# (5 - 2.857)^2 / 2.857 ≈ 1.617
# (3 - 2.857)^2 / 2.857 ≈ 0.007
# (0 - 2.286)^2 / 2.286 ≈ 2.287
# (0 - 2.143)^2 / 2.143 ≈ 2.143
# (2 - 2.143)^2 / 2.143 ≈ 0.010
# (4 - 1.714)^2 / 1.714 ≈ 3.097

# Total chi-square ≈ 9.161
# Degrees of freedom = (2-1)*(3-1) = 2
# At α = 0.05, critical value from chi-square table = 5.991
# Since 9.161 > 5.991, result is significant.

# Conclusion: ✅ There is a significant association between Age and Movie preference.
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# =============================================
# Dataset 1: NON-SIGNIFICANT Relationship
# Gender vs. Beverage Preference
# =============================================
data1 = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Beverage': ['Coffee', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Coffee', 'Tea', 'Tea']
}
df1 = pd.DataFrame(data1)

cont_table1 = pd.crosstab(df1['Gender'], df1['Beverage'])
print("Dataset 1 - No Significant Relationship:")
print(cont_table1)
print("\n")

chi2_1, p1, dof1, expected1 = chi2_contingency(cont_table1)
print(f"Chi2 Statistic: {chi2_1:.4f}")
print(f"P-value: {p1:.4f}")
print("Expected Frequencies:")
print(expected1)
print("\n" + "="*50 + "\n")

# =============================================
# Dataset 2: SIGNIFICANT Relationship
# Age Group vs. Movie Preference
# =============================================
data2 = {
    'Age': ['Young', 'Young', 'Old', 'Old', 'Young', 'Young', 'Old', 'Old',
            'Young', 'Old', 'Old', 'Old', 'Young', 'Young'],
    'Movie': ['Action', 'Comedy', 'Comedy', 'Drama', 'Action', 'Action',
              'Drama', 'Drama', 'Action', 'Drama', 'Drama', 'Comedy', 'Action', 'Comedy']
}
df2 = pd.DataFrame(data2)

cont_table2 = pd.crosstab(df2['Age'], df2['Movie'])
print("Dataset 2 - Significant Relationship:")
print(cont_table2)
print("\n")

chi2_2, p2, dof2, expected2 = chi2_contingency(cont_table2)
print(f"Chi2 Statistic: {chi2_2:.4f}")
print(f"P-value: {p2:.4f}")
print("Expected Frequencies:")
print(expected2)
print("\n" + "="*50 + "\n")

# =============================================
# Visualization using Bar Graphs
# =============================================
plt.figure(figsize=(12, 5))

# Bar plot for Dataset 1 (No Significant Relationship)
plt.subplot(1, 2, 1)
cont_table1.plot(kind='bar', stacked=False, color=['blue', 'green'], ax=plt.gca())
plt.title("No Significant Relationship\n(Gender vs. Beverage)")
plt.ylabel('Frequency')
plt.xlabel('Gender')


# Bar plot for Dataset 2 (Significant Relationship)
plt.subplot(1, 2, 2)
cont_table2.plot(kind='bar', stacked=False, color=['red', 'yellow', 'green'], ax=plt.gca())
plt.title("Significant Relationship\n(Age vs. Movie Preference)")
plt.ylabel('Frequency')
plt.xlabel('Age Group')

plt.tight_layout()
plt.show()
```
